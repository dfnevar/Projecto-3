# Importo las librerias 

```python
import pandas as pd
import seaborn as sns
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

# Importo los datasets

```python
df = pd.read_csv('https://opendata-ajuntament.barcelona.cat/data/dataset/69c3250b-100c-4d5b-a08e-0a191d9950e3/resource/e96bf614-467b-40ab-91b9-e48a616ea775/download/2022_lloguer_preu_trim.csv')
df2 = pd.read_csv('https://opendata-ajuntament.barcelona.cat/data/dataset/69c3250b-100c-4d5b-a08e-0a191d9950e3/resource/cfc45f2b-62eb-4621-8486-1b90e36b4bfe/download/2021_lloguer_preu_trim.csv')
df3 = pd.read_csv('https://opendata-ajuntament.barcelona.cat/data/dataset/69c3250b-100c-4d5b-a08e-0a191d9950e3/resource/47c9d64d-317a-45d0-8c45-45488df8601c/download/2020_lloguer_preu_trim.csv')

# CONCATENATO LOS 3 DATA SET PARA TENER UN UNICO ARCHIVO 
df = pd.concat([df, df2, df3])
```

# Elimino los filas, columnas o datos que no son relevantes 

```python
def trimestreAMes(x):
    """
    Como los datos se registran en trimestres y necesitamos fechas para plottearlo mejor, transformamos
    trimestres (1,2,3,4) a meses (1..12)
    """
    if x == 1:
        return 1
    elif x == 2:
        return 4
    elif x == 3:
        return 7
    elif x == 4:
        return 10

    
def preprocessing(df):
    df.drop(df[df['Lloguer_mitja'] == 'Lloguer mitjà per superfície (Euros/m2 mes)'].index, inplace=True)
    df.dropna(inplace=True)
    df["mes"] =  df["Trimestre"].apply(trimestreAMes)
    df["fecha"] = pd.to_datetime(df["Any"].astype(str) + "-" + df["mes"].astype(str) , format='%Y-%m')
    df.loc[df["Preu"] == "--", "Preu"] = np.nan
    df.dropna(inplace=True)
    df["Preu"] = df["Preu"].astype(float)
    return df
```

```python
df = preprocessing(df)
df
```

```python
def plotMinMaxMeanByDistrict(df):
    # lo mismo que hice arriba pero le aplico el max min y mean todo al mismo tiempo
    df1 = df.groupby(["Nom_Districte", "fecha"])["Preu"].agg(["min", "max", "mean"]).reset_index()
    g = sns.FacetGrid(df1, col="Nom_Districte", col_wrap=4, height=3, aspect = 1)
    g.map(sns.lineplot, "fecha", "min")
    g.map(sns.lineplot, "fecha", "max", **{"color" : "red"}) 
    g.map(sns.lineplot, "fecha", "mean", **{"color" : "green"})

    g.set_ylabels("Price in euros")
    g.fig.suptitle("Max (red), min(blue) and mean(green) price evolution by district")
    g.fig.subplots_adjust(top=0.9)

    for axes in g.axes.flat:
        _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=45)

plotMinMaxMeanByDistrict(df)

plt.savefig('img4.jpg')
```

```python
# grafico la media de precios en toda la ciudad
meanPriceByDate = df.groupby("fecha").mean()

sns.relplot(data=meanPriceByDate, x="fecha", y="Preu", kind="line")
plt.xticks(rotation=45)
plt.show()

plt.savefig('img6.jpg')
```

```python
#prueba estadistica para ver si la serie es estacionaria o no 
from statsmodels.tsa.stattools import adfuller
ts_data = df.groupby("fecha").mean()["Preu"].values

#aplico la prueba 
result = adfuller(ts_data, maxlag = 3)

#imprimo resultados de la prueba
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
    
#si fuera mayor a 0.05 fayo en rechazara la hipotesis nula 
#por lo que podriamos decir que los datos estan en una serie no estacionaria 
```

# Porcentaje de incremente desde el 2020/01 al 2022/12

```python
# este es el porcetnaje que subio la media desde la primera fecha tomada hasta la ultima 
100 * (df.groupby("fecha").median()["Preu"].values[-1] - df.groupby("fecha").median()["Preu"].values[0]) / df.groupby("fecha").median()["Preu"].values[0]
```

```python
def getIncremento(df):
    """
    Devuelve el porcentaje en que cambiaron los precios en cada distritio desde la primera fecha a analizar
    """
    vals = df.groupby(["Nom_Districte", "fecha"]).median()["Preu"].reset_index()
    # en esta lista meti el distrito y el porcentaje de cuanto a subido 
    incremento = []
    for districte in df["Nom_Districte"].unique():
        vals_districte = vals[vals["Nom_Districte"] == districte]
        fecha_max_val = vals_districte[vals_districte["fecha"] == vals_districte["fecha"].max()]
        fecha_min_val = vals_districte[vals_districte["fecha"] == vals_districte["fecha"].min()]

        incremento.append((districte, 
            100 * (fecha_max_val["Preu"].values[0] - fecha_min_val["Preu"].values[0]) / fecha_min_val["Preu"].values[0])
        )
    return incremento
```

```python
incremento = getIncremento(df)
incremento
```

```python
df_aux = pd.DataFrame(incremento, columns=["Nom_Districte", "Incremento"])

df_aux.plot(kind="bar", x="Nom_Districte", y="Incremento")

plt.savefig('img4.jpg')
```
